```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
result=[]
loc=[]
shop_name=[]
shop_loc=[]
shop_num=[]
for i in range(1, 54): 
    html=urlopen('https://www.hollys.co.kr/store/korea/korStore2.do?pageNo='+str(i)+'&sido=&gugun=&store=')
    soup=BeautifulSoup(html, 'html.parser')
    TB=soup.find_all('tbody')
    
    for i in TB:
        td=i.find_all('td')
        for i in td:
            result.append(i.text)
                   
for i in range(0, len(result)):
    if i % 6 == 0:
        loc.append(result[i])
    if i % 6 == 1:
        shop_name.append(result[i])
    if i % 6 == 3:
        shop_loc.append(result[i])
    if i % 6 == 5:
        shop_num.append(result[i])
    
loc=pd.DataFrame(loc)
shop_name=pd.DataFrame(shop_name)  
shop_loc=pd.DataFrame(shop_loc)  
shop_num=pd.DataFrame(shop_num)  

    
df=pd.concat([loc, shop_name, shop_loc, shop_num], axis=1)
df.columns=['지역', '매장명', '주소', '전화번호']

list=[]
for i in range(0, len(shop_name)):
    print('[ {0}]매장이름: {1}, 지역: {2}, 주소: {3}, 전화번호: {4}'.format(i+1, df.iloc[i]['매장명'], df.iloc[i]['지역'], df.iloc[i]['주소'], df.iloc[i]['전화번호']))
print(f'전체 매장 수: {df.shape[0]}')
df.to_csv('hollys_branches.csv', encoding='utf-8')
            
     
```
